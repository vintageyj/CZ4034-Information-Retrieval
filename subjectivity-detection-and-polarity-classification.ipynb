### We give a threshold of -0.05 to 0.05


In the code example provided earlier, I used a threshold of 0.05 (and -0.05 for negative sentiment) to classify the sentiment into positive, negative, or neutral categories. This threshold is somewhat arbitrary and can be adjusted based on the specific requirements of your application or the characteristics of your data.

Here's why I chose 0.05 instead of 0:

Neutral Sentiment: Sentiment analysis often deals with texts that may not express strong sentiment in either direction. By using a threshold slightly above or below 0, we can classify sentiments closer to 0 as neutral. This allows for a more balanced classification, distinguishing between slightly positive, slightly negative, and neutral sentiments.

Data Variability: The choice of threshold may depend on the variability and characteristics of your data. In some datasets, sentiment scores may vary widely, and using a threshold closer to 0 may result in a large number of neutral classifications. Adjusting the threshold allows you to fine-tune the classification based on the specific characteristics of your data.

Application Specifics: The choice of threshold may also depend on the specific application or task you're working on. For example, in some applications, you may want to be more conservative in classifying sentiment and only label texts with strong sentiment as positive or negative.

### TextBlob

```python
blob12345 = TextBlob("This is just a somewhat decent sentence")
print(blob12345.sentiment.subjectivity)
print(blob12345.sentiment.polarity)
```

```python
from textblob import TextBlob

# Predict subjectivity and sentiment for each sentence
for sentence in tricky_sentences:
    blob = TextBlob(sentence)
    subjectivity = blob.sentiment.subjectivity
    sentiment = blob.sentiment.polarity
    
    # Classify sentiment
    if sentiment > 0.05:
        sentiment_label = "Positive"
    elif sentiment < -0.05:
        sentiment_label = "Negative"
    else:
        sentiment_label = "Neutral"
    
    print(f"Sentence: '{sentence}'")
    print(f"Subjectivity: {subjectivity}")
    print(f"Sentiment: {sentiment} ({sentiment_label})")
    print()
```

### VADER from NLTK

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER lexicon
nltk.download('vader_lexicon')

# Initialize VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Predict subjectivity and sentiment for each sentence
for sentence in tricky_sentences:
    scores = sid.polarity_scores(sentence)
    print(scores)
    sentiment = scores['compound']
    
    # Classify sentiment
    if sentiment > 0.05:
        sentiment_label = "Positive"
    elif sentiment < -0.05:
        sentiment_label = "Negative"
    else:
        sentiment_label = "Neutral"
    
    print(f"Sentence: '{sentence}'")
    print(f"Sentiment: {sentiment} ({sentiment_label})")
    print()
```

```python
def sentiment_score_1(review):
    tokens = tokenizer_1.encode(review, return_tensors='pt')
    result = model_1(tokens)
    prediction = int(torch.argmax(result.logits))

    # Classify sentiment
    if prediction == 1:
        sentiment_label = "Positive"
    else:
        sentiment_label = "Negative"
    
    return sentiment_label # return logits as well?
```

```python
# Import dataset, using first column as index column
glassdoor_data_textblob = pd.read_csv('sg_companies_reviewsYJ.csv')

# Concatenating Pros and Cons
glassdoor_data_textblob['Overall Review'] = glassdoor_data_textblob['Pros'] + '. ' + glassdoor_data_textblob['Cons']
```

```python
# Create empty lists to store data
subjectivities = []
subjectivity_labels = []
sentiments = []
sentiment_labels = []

# Predict subjectivity and sentiment for each sentence
for review in glassdoor_data_textblob['Overall Review']:
    blob = TextBlob(review)
    subjectivity = blob.sentiment.subjectivity
    sentiment = blob.sentiment.polarity
    
    # Classify sentiment
    if sentiment > 0.05:
        sentiment_label = "Positive"
    elif sentiment < -0.05:
        sentiment_label = "Negative"
    else:
        sentiment_label = "Neutral"
    
    # Detect subjectivity
    if subjectivity > 0.7:
        subjectivity_label = "Highly Opinionated sentence"
    elif subjectivity > 0.2:
        subjectivity_label = "Less Opinionated sentence"
    else:
        subjectivity_label = "Objective sentence"
    
    # Append the results to the lists
    subjectivities.append(subjectivity)
    subjectivity_labels.append(subjectivity_label)
    sentiments.append(sentiment)
    sentiment_labels.append(sentiment_label)
    
# Adding new columns to glassdoor_data_textblob
glassdoor_data_textblob['Subjectivity TextBlob'] = np.array(subjectivities)
glassdoor_data_textblob['Subjectivity Label TextBlob'] = np.array(subjectivity_labels)
glassdoor_data_textblob['Predicted Sentiment TextBlob'] = np.array(sentiments)
glassdoor_data_textblob['Predicted Label TextBlob'] = np.array(sentiment_labels)
```
