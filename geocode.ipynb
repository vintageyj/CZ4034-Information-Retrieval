```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

def get_continent(latitude, longitude):
    geolocator = Nominatim(user_agent="geoapiExercises")
    try:
        # Reverse geocode to get the location details from (latitude, longitude)
        location = geolocator.reverse((latitude, longitude), exactly_one=True)
        if location:
            # Parse location raw data for the continent
            address = location.raw.get('address', {})
            continent = address.get('continent', 'Continent not found')
            return continent
    except GeocoderTimedOut:
        return "GeocoderTimedOut: Service timed out"

# Example usage
latitude, longitude = 42.222261,	-83.396599  # Coordinates for Paris, France
continent = get_continent(latitude, longitude)
print(f"The continent is: {continent}")
```

```python
from geopy.geocoders import GoogleV3
from geopy.exc import GeocoderTimedOut

def get_country(latitude, longitude, api_key):
    if pd.isna(latitude) or pd.isna(longitude):
        return np.nan
    geolocator = GoogleV3(api_key=api_key)
    try:
        # Reverse geocode to get the location details from (latitude, longitude)
        location = geolocator.reverse((latitude, longitude), exactly_one=True)
        if location:
            # Iterate through the address components to find the country
            for component in location.raw['address_components']:
                if 'country' in component['types']:
                    return component['long_name']  # Returns the full name of the country
            return "Country not found"
    except GeocoderTimedOut:
        return "GeocoderTimedOut: Service timed out"

# Replace 'YOUR_API_KEY_HERE' with your actual Google Maps Geocoding API key
api_key = ''

# Example usage with coordinates for Paris, France
latitude, longitude = 48.8566, 2.3522
country = get_country(latitude, longitude, api_key)
print(f"The country is: {country}")
```
