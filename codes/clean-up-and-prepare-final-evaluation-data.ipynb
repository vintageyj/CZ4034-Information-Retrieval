### Filter rows that had changes:

```python
result = raw_evaluation_data[raw_evaluation_data['Check'].notna()]

result
```

```python
result.shape
```

```python
indexes_to_remove = result.index
indexes_to_remove
```

### 42 of them were not sure or labelled wrongly. Remove them, should have enough data for evaluation:

```python
# Remove rows with indexes in the list
raw_evaluation_data_cleaned = raw_evaluation_data.drop(indexes_to_remove)

raw_evaluation_data_cleaned
```

```python
raw_evaluation_data_cleaned.shape
```

```python
raw_evaluation_data_cleaned['Sentiment'].value_counts()
```

### Undersample Positive class to get 50:50 ratio:

```python
# Set a random seed for reproducibility
np.random.seed(42)
```

```python
# Count the number of positive and negative samples
positive_count = raw_evaluation_data_cleaned['Sentiment'].value_counts()['Positive']
negative_count = raw_evaluation_data_cleaned['Sentiment'].value_counts()['Negative']

# Determine the desired number of samples from each class
desired_count = min(positive_count, negative_count)

# Undersample the positive class
undersampled_positive = raw_evaluation_data_cleaned[raw_evaluation_data_cleaned['Sentiment'] == 'Positive'].sample(desired_count)

# Keep all negative samples
negative_samples = raw_evaluation_data_cleaned[raw_evaluation_data_cleaned['Sentiment'] == 'Negative']

# Concatenate the undersampled positive samples with all negative samples
balanced_df = pd.concat([undersampled_positive, negative_samples])

balanced_df
```
